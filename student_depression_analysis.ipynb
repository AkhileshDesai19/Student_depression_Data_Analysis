## Problem Statement


Student depression has become a growing concern worldwide, affecting academic performance, mental well-being, social relationships, and overall quality of life. Various factors such as academic pressure, study satisfaction, financial stress, sleep habits, age, gender, and family background may contribute to depression among students.

The objective of this project is to perform Exploratory Data Analysis (EDA) on the Student Depression Dataset to identify patterns, trends, and relationships between different student attributes and depression levels. Through data visualization and statistical analysis, the project aims to uncover key factors associated with student depression and provide meaningful insights that can help educational institutions, counselors, and policymakers better understand student mental health challenges.


### Business Objective

The primary objective of this analysis is to understand the factors that influence depression among students and identify the characteristics most strongly associated with depressive symptoms.

Specific objectives include:

* Analyzing the distribution of depression among students.
* Examining the impact of academic pressure on student depression.
* Studying the relationship between study satisfaction and depression.
* Investigating how sleep duration affects mental health.
* Understanding the influence of financial stress on depression levels.
* Exploring whether demographic factors such as age and gender are associated with depression.
* Identifying key patterns and trends that can help educational institutions design effective mental health support programs.
* Providing data-driven insights that can assist counselors, universities, and policymakers in improving student well-being and academic success.

### Univariate Analysis for Numeric Columns 

```python
num_col = ['Age', 'CGPA', 'Work/Study Hours']

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for i, col in enumerate(num_col):
    sns.histplot(df[col], bins=20, kde=True, ax=axes[i])
    axes[i].set_title(col)

plt.tight_layout()
plt.show()
```

## Insight 
- Most students are between 18 and 35 years old, with no major age outliers.

- CGPA is concentrated in the 6–10 range, indicating relatively good academic performance among respondents.

- Work/Study Hours are concentrated toward the higher end (8–12 hours), suggesting many respondents spend substantial time studying or working.

## Categorical Columns 

```python

cat_col = [
    'Gender',
    'Sleep Duration',
    'Dietary Habits',
    'Have you ever had suicidal thoughts ?',
    'Family History of Mental Illness',
    'Depression'
]

fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.flatten()

for i, col in enumerate(cat_col):

    order = df[col].value_counts().index

    ax = sns.countplot(
        x=col,
        data=df,
        order=order,
        ax=axes[i],

    )

    total = len(df)

    for p in ax.patches:
        percentage = f'{100 * p.get_height() / total:.1f}%'

        ax.text(
            p.get_x() + p.get_width()/2,
            p.get_height(),
            percentage,
            ha='center',
            va='bottom'
            
            
        )

    ax.set_title(col)
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python

to_cat = [
    'Academic Pressure',
    'Financial Stress',
    'Work Pressure',
    'Study Satisfaction',
    'Job Satisfaction'
]

# Adjust according to your actual labels
order = ['Very Low', 'Low', 'Moderate', 'Average', 'High', 'Very High']

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, col in enumerate(to_cat):

    ax = sns.countplot(
        data=df,
        x=col,
        order=[x for x in order if x in df[col].unique()],
        ax=axes[i]
    )

    total = len(df)

    for p in ax.patches:
        pct = 100 * p.get_height() / total

        ax.annotate(
            f'{pct:.1f}%',
            (p.get_x() + p.get_width()/2, p.get_height()),
            ha='center',
            va='bottom'
        )

    ax.set_title(col)
    ax.tick_params(axis='x', rotation=45)

# Remove the extra empty subplot
if len(to_cat) < len(axes):
    fig.delaxes(axes[-1])

plt.tight_layout()
plt.show()
```
